```
import pandas as pd

county_cases_data = pd.read_csv("https://rajstorage11.blob.core.windows.net/test/us-counties.csv")
unemployment_data = pd.read_csv("https://rajstorage11.blob.core.windows.net/test/Unemployment.csv", index_col="FIPStxt")
county_area_data = pd.read_csv("https://rajstorage11.blob.core.windows.net/test/Counties.csv", index_col="GEOID")
population_data = pd.read_csv("https://rajstorage11.blob.core.windows.net/test/PopulationEstimates.csv", index_col="FIPStxt", encoding='latin-1')
```

```
population_data.head()
```

```
county_area_data.head()
```

```
population_data = population_data.rename(columns={'POP_ESTIMATE_2019':'population'})
population_series = population_data['population']
population_series
```

```
county_area_data = county_area_data.rename(columns={'Shape.STArea()':'area'})
area_series = county_area_data['area']
area_series = area_series.sort_index()
area_series
```

```
combined_data = population_series.to_frame().join(area_series)
combined_data.head()
```

```
combined_data = combined_data.dropna()
combined_data.head()
```

```
def my_function(some_text):
  return some_text.replace(',' , '')

combined_data['population'] = combined_data['population'].apply(my_function)
combined_data.head()

```

```
combined_data['population'] = combined_data['population'].astype('int')
combined_data.head()
```

```
combined_data['population density'] = combined_data['population'] / combined_data['area']
combined_data.head()
```
